```python
#packages required
from tkinter import *
from tkinter import ttk
from tkinter.font import Font
from tkinter import messagebox 
import pandas as pd
import numpy as np
from heapq import nlargest
```

```python
#The classifier model is imported here
import pickle
f=open('yoga_therpy.pickle','rb')
classifier=pickle.load(f)
f.close() 
```

```python
#The data file
med_data=pd.read_csv('medical_data.csv')
med_data.fillna(0, inplace=True)
med_data.head()
```

```python
#validation dataset
validation_df = med_data.iloc[0:1,1:]
validation_df
```

```python
yoga_rec = pd.read_csv('yoga_practices.csv', index_col = 'disease')
yoga_rec #consists of the mapping of diseases with the yoga practices recommended
```

```python
yoga_rec.description['asthma']
```

```python
yoga_dis = yoga_rec.index
yoga_dis #list of diseases
```

```python
yoga_dis.get_loc('influenza')
```

```python
data_var = pd.read_csv('medical_data.csv', index_col=0)
data_var.head(10)
#getting the symptoms list
symptoms_dis = list(data_var.columns[1:])
symptoms_dis = sorted(symptoms_dis) #Sort the list of symptoms
symptoms_dis = symptoms_dis[2:]
symptoms_dis  
```

```python
#Creation of the UI
app = Tk()  #App's main Window
app.geometry("4000x1000")
app.title("New App")
Label(app, text = "Welcome to Yoga Therapy Recommendation!!!!",fg = 'brown', font = ('Times New Roman',30)).pack()
```

```python
List = [] #the validation data is passed to this list

#Click Here button Handle    
def clickMe():
    list_action.config(state = NORMAL)
    #Setting the font!
    T.configure(state = NORMAL)
    defFont1 = Font(family="Times New Roman", size=13)
    T.tag_configure('default_font',font=defFont1,foreground = 'black')
    T.configure(state = DISABLED)
    
    #getting the variable and displaying in the textbox!            
    dis = variable.get()
    print(dis)
    if dis in symptoms_dis:    #if only the symptoms are listed in the combobox it is dispayed in the textbox
        disease_ob = dis
        #print(disease_ob)
    T.configure(state = NORMAL)
    T.insert(INSERT,disease_ob.title() + '\n',"default_font")
    List.append(variable.get())
    T.configure(state = DISABLED)
    
#Disease button Handle     
def clickOK():
    T.configure(state = NORMAL)
    T.delete(1.0,END) #clear the display
    T.configure(state = DISABLED)
    action.config(state = DISABLED)
    list_action.config(state = DISABLED)
    print(List)
    #Set the validaton table!            
    col_lst=validation_df.columns
    for i in col_lst:
        validation_df[i]=0
    for i in List:
        validation_df[i]=1
    #predicting the value
    pred_val=classifier.predict(validation_df) #Fisrt prediction
    print(pred_val)
    prob = classifier.predict_proba(validation_df) #probabilities
    prob1=(nlargest(3,prob[0])) #top three probable values
    prob2 = [] #Final list
    #condition to print only the most probaable diseases
    for x in prob1:
        if x>0.2:
            prob2.append(x)
    print(prob[0])
    print(prob1)
    print(prob2)
    global num #total number of elements
    num = len(prob[0]) 
    print(num)
    global ind #list of indices of the probable values
    ind = [] 
    predictions_val = [] #to hold the list of probable values with 20-50% confidence
    #global yogarec
    #yogarec = [] #list to hold the yoga recommendation values
    if (len(prob2)>0):
        if (prob2[0]<=0.5):
            for i in range(num):
                if prob[0][i] in prob2:
                    predictions_val.append(prob[0][i])
                    ind.append(i)
                else:
                    predictions_val.append(0)
            defFont = Font(family="Times New Roman", size=13)
            T.configure(font=defFont , foreground = 'blue')
            myFont = Font(family="Times New Roman", size=12, weight="bold", slant="italic")
            T.tag_configure("bold1", font=myFont, justify='center',foreground = 'red')
            T.configure(state = NORMAL)                        
            T.insert(INSERT,"The most probable diseases found are:\n")
            
            for i in ind:
                T.insert(INSERT,yoga_dis[i].title() + '\n', "bold1")
                print(yoga_dis[i])
                #yogarec.append(yoga_rec.description[yoga_dis[i]])
            yoga_recom.config(state = NORMAL)
            List.clear() #clear the list so that next list can be generated from selections
            T.configure(state = DISABLED)
        else:
            pred_val = str(pred_val[0])
            print(pred_val)
            ind.append(yoga_dis.get_loc(pred_val))
            print(ind)
            #setting fonts and configuring
            defFont = Font(family="Times New Roman", size=13)
            T.configure(font=defFont , foreground = 'blue')
            myFont = Font(family="Times New Roman", size=12, weight="bold", slant="italic")
            T.tag_configure("bold1", font=myFont, justify='center',foreground = 'red')
            T.configure(state = NORMAL)                        
            T.insert(INSERT,"The disease has been found to be\n")
            T.insert(INSERT,pred_val.title(), "bold1")
            yoga_recom.config(state = NORMAL)
            List.clear() #clear the list so that next list can be generated from selections
            T.configure(state = DISABLED)
            #yogarec.append(yoga_rec.description[pred_val])
            return
    else:
        T.configure(state = NORMAL)
        T.delete(1.0,END) #clear the display
        myFont = Font(family="Times New Roman", size=12, weight="bold", slant="italic")
        T.tag_configure("bold1", font=myFont, justify='center',foreground = 'red')  
        T.insert(INSERT, "No matching diseases found!!!", "bold1")
        T.configure(state = DISABLED)
        action.configure(state = NORMAL)
        app.after(1000,pass2)
        return 
#To return back to the selection pane
def pass2():
    T.configure(state = NORMAL)
    T.delete(1.0,END) #clear the display
    T.configure(state = DISABLED)
    variable.set("Select the matching symptoms..")
    action.config(state=NORMAL)
    yoga_recom.configure(state = DISABLED)
    List.clear()
    
#Yoga Recommendation button Handle     
def recommend():
    #T.delete(1.0,END)
    #T.configure(font=myFont)
    myFont1 = Font(family="Times New Roman", size=13, weight="bold", slant="italic")
    T.tag_configure("bold1", font=myFont1, justify='left',foreground = 'red')  
    myFont = Font(family="Times New Roman", size=12, weight="bold", slant="italic")
    T.tag_configure("bold", font=myFont, justify='left',foreground = 'green')
    #text.insert("end", "Hello, ") 
    #text.insert("end", "world", "bold") 
    T.configure(state = NORMAL)
    T.delete(1.0,END) #clear the display
    T.insert(INSERT,"The recommended yoga exercises for the disease are:  \n")
   
    for i in ind:
        T.insert(INSERT,yoga_dis[i].title() + '\n', "bold1")
        T.insert(INSERT,yoga_rec.description[i] + '\n', "bold")
    T.configure(state = DISABLED)
    #print(yogarec)
    app.after(1000,pass1)
    #yogarec.clear()
    
#Message box Handle
def pass1():
    if messagebox.askyesno('Confirm', 'Do you want to Continue?'):
        T.configure(state = NORMAL)
        T.delete(1.0,END)
        T.configure(state = DISABLED)
        variable.set("Select the matching symptoms..")
        action.config(state=NORMAL)
        yoga_recom.configure(state = DISABLED)
    else:
        if messagebox.askyesno('Verify', 'Do you want to quit?'):
            app.destroy()
        else:
            T.configure(state = NORMAL)
            T.delete(1.0,END)
            T.configure(state = DISABLED)
            action.config(state=NORMAL)
            variable.set("Select the matching symptoms..")
            yoga_recom.configure(state = DISABLED)
            
#Exit button handle
def ExitMe():
    if messagebox.askyesno('Verify', 'Really quit?'):
        app.destroy()
    else:
        T.configure(state = NORMAL)
        T.delete(1.0,END)
        T.configure(state = DISABLED)
        action.config(state=NORMAL)
        variable.set("Select the matching symptoms..")
        yoga_recom.configure(state = DISABLED)   
```

```python
ttk.Label(app, text="Select the symptoms from the drop down and click the button..",foreground = 'dark blue', font = "Times").pack(padx = 40)

#variables for the combobox
variable = StringVar(app)
variable.set("Select the matching symptoms..") # default value

#Initialising the combobox
w = ttk.Combobox(app, state="readonly", textvariable=variable, values=symptoms_dis, width = 30 )
w.pack(padx = 500, pady = 20, fill = X, side = TOP)

#Click Here Button        
action = ttk.Button(app, text="Click Here!", command=clickMe)
action.place(x = 1050, y = 93)

#Textbox initialisation
T = Text(app, height=10, width=80, state = DISABLED)
T.configure(state = DISABLED)
T.pack()

# button fr disease
list_action = ttk.Button(app, text="Disease",state = DISABLED, command=clickOK)
list_action.place(x = 620, y = 330)

# button fr yoga recommendation
yoga_recom = ttk.Button(app, text="Yoga Recommended",state = DISABLED, command=recommend)
yoga_recom.place(x = 700, y = 330)

#Exit button
exit_bt = ttk.Button(app, text="Exit", command=ExitMe)
exit_bt.place(x = 826, y = 330)
```

```python
app.mainloop()
```
